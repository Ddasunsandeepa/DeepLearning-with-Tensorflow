```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('sonar.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
x = df.iloc[:, 0:60].astype(float)
y = df.iloc[:, 60]
```

```python
x.shape
```

```python
encorder = LabelEncoder()
encorder.fit(y)
y = encorder.transform(y)
```

```python
y
```

```python
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
tf.random.set_seed(42)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(30, activation='relu', input_shape=(60,)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(60, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(68, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(38, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation ='sigmoid')
])
```

```python
model.compile(loss = tf.keras.losses.BinaryCrossentropy(),
optimizer = tf.keras.optimizers.Adam(learning_rate=0.005),
metrics = ['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```
